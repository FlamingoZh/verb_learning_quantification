```python
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import numpy as np
import pandas as pd
```

```python
save_fig=True
```

```python
# sns.set_style('darkgrid')  # darkgrid, white grid, dark, white and ticks
plt.rc('axes', titlesize=18)  # fontsize of the axes title
plt.rc('axes', labelsize=18)  # fontsize of the x and y labels
plt.rc('xtick', labelsize=18)  # fontsize of the tick labels
plt.rc('ytick', labelsize=18)  # fontsize of the tick labels
plt.rc('legend', fontsize=18)  # legend fontsize
plt.rc('font', size=18)  # controls default text sizes
# plt.figure(figsize=(10, 6), tight_layout=True)
```

```python
def unpack_data(data):
    n_exemplar_list = data["n_exemplar_list"]
    y_list=data["y_list"]
    datasetname = data["datasetname"]
    n_exemplar_max = data["n_exemplar_max"]
    n_sample = data["n_sample"]
    aggregation_mode = data["aggregation_mode"]
    extra_info=data["extra_info"]
    return n_exemplar_list[1:],y_list[1:]
```

```python
vg_noun_data_path="../data/dumped_plot_data/vg_noun_bert_visual_20_100.pkl"
vg_noun_data=unpack_data(pickle.load(open(vg_noun_data_path,"rb")))

vg_verb_data_path="../data/dumped_plot_data/vg_verb_bert_visual_20_100.pkl"
vg_verb_data=unpack_data(pickle.load(open(vg_verb_data_path,"rb")))

df1 = pd.DataFrame(np.array(vg_noun_data).transpose(), columns = ['x', 'y'])
df1['word_type']="Nouns"

df2 = pd.DataFrame(np.array(vg_verb_data).transpose(), columns = ['x', 'y'])
df2['word_type']="Verbs"

df_visual=pd.concat([df1,df2],ignore_index=True)
```

```python
fig, ax = plt.subplots(1, figsize=(10, 8))
palette=[sns.color_palette("gist_stern_r")[i] for i in [0,5]]
sns.pointplot(x="x",y="y",data=df_visual,hue="word_type",palette=palette,ax=ax)

# Remove the top and right spines from plot
sns.despine()

ax.legend(loc=4)
ax.set_xlim(-1, 21)
ax.set_xticks([-1,4,9,14,19])
ax.set_xticklabels([0,5,10,15,20])
ax.set_ylim(0.73, 1.02)
ax.set_yticks([0.75,0.8,0.85,0.9,0.95,1.0])

ax.set_xlabel("Number of Visual Exemplars Aggregated",fontname='Georgia',fontsize=24)
ax.set_ylabel("Relative Alignment Strength",fontname='Georgia', fontsize=24)

ax.text(6,1.03,"Visual Aggregation",fontname='Georgia', fontsize=28)

plt.tight_layout()
if save_fig:
    plt.savefig("../figs/visual_agg_vg.png",dpi=600)
plt.show()
```
